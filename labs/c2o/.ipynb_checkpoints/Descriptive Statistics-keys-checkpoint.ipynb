# Descriptive Statistics
The market research team at AdRight is assigned the task to identify the profile of the typical customer for each treadmill product offered by CardioGood Fitness. The market research team decides to investigate whether there are differences across the product lines with respect to customer characteristics. 

### The team identifies the following customer variables to study: 
  - product purchased, TM195, TM498, or TM798; 
  - gender; 
  - age, in years; 
  - education, in years; 
  - relationship status, single or partnered; 
  - annual household income ; 
  - average number of times the customer plans to use the treadmill each week; 
  - average number of miles the customer expects to walk/run each week; 
  - and self-rated fitness on an 1-to-5 scale, where 1 is poor shape and 5 is excellent shape.

### Task: Perform descriptive analytics to create a customer profile for each treadmill product line.

```python
# Load the necessary packages

import numpy as np
import pandas as pd
```

```python
# Load the Cardio Dataset
mydata = pd.read_csv('data/CardioGoodFitness.csv')
```

```python
#display data
mydata.head(10)
```

```python
#Describe the data
mydata.describe(include="all")
```

```python
#get info about the data for each colum and its types
mydata.info()
```

```python
# Print the min, max and Standard Deviation for any one column

print(mydata['Age'].std())
print(mydata['Age'].min())
print(mydata['Age'].max())
```

```python
# Plot Histograms
# Age 
# Income, etc

import matplotlib.pyplot as plt

mydata.hist(figsize=(20,30))
```

```python
# Group Product by gender 
pd.crosstab(mydata['Product'],mydata['Gender'] )
```

```python
# Group Product by MaritalStatus
pd.crosstab(mydata['Product'],mydata['MaritalStatus'] )
```

```python
import seaborn as sns
```

```python
# A bar plot to show the products broken down by gender count.

sns.countplot(x="Product", hue="Gender", data=mydata)
```

### Creating summaries of the dataset

```python
# Explore Pandas Pivot Table feature

# A pandas pivot table has three main elements. The index specifies the row-level grouping, columns specify the column level grouping and values which are the numerical values you are looking to summarise.


ProductGender_by_MaritalStatus=pd.pivot_table(mydata, index=['Product', 'Gender'],columns=[ 'MaritalStatus'], aggfunc=len)
```

```python
ProductGender_by_MaritalStatus
```

```python
#Find Income Distribution by grouping Product, Gender for each of the values in MaritalStatus

pd.pivot_table(mydata,'Income', index=['Product', 'Gender'], columns=[ 'MaritalStatus'])
```

```python
#Find Miles Covered by grouping Product, Gender for each of the values in MaritalStatus

pd.pivot_table(mydata,'Miles', index=['Product', 'Gender'], columns=[ 'MaritalStatus'])
```

```python
# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Create a visualization
sns.relplot(
    data=mydata,
    x="Income", y="Product", col="Gender",
    hue="MaritalStatus", style="MaritalStatus",
)
```

```python
#  Using seaborn library Explore other vizualizations and pairplots.
sns.pairplot(mydata)
plt.show()
#  Share any interesting Relationships and Findings with class
```
